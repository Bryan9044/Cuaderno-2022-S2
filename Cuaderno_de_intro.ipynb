## Contador de digitos

Usando Pseint crear un diagrama de flujo para representar el flujo y pseudocodigo para el algoritmo de contarDigitos

-Recibiran dos parametros 
-Ambos mayores a cero



### Diagrama de flujo 
image.png


### Pseudocodigo 


```
Algoritmo contarDigitos
	total <- 0
	Leer a
	Mientras a>0 Hacer
		// Realiza la operacion
		a <- trunc(a/10)
		total <- total+1
	FinMientras
	Escribir total
FinAlgoritmo

```
