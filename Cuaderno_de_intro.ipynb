Usando Pseint crear un diagrama de flujo para representar el flujo y pseudocodigo
- Se recibe un parametro el cual es num que sera al que le vamos a sacar el factorial y luego num1 que sera por el cual vamos a multiplicar nuestro num para dar con el resultado final



image.png

Pseudocodigo


```
Algoritmo Factorial
	Leer num
	Si num>=0 Entonces
		num1 = 1
		Mientras num >1 Hacer
			num1 = num1 * num
			num = num - 1
		FinMientras
		Escribir num1
	FinSi
FinAlgoritmo

```



##Codigo


```
def factorial(num):
    num2 = 1
    if isinstance(num, int):
        if num > 0:
            while num > 1:
                num2 = num2 * num
                num = num - 1
            return num2
        else:
            return"Error: el num no puede ser menor a CERO"
    else:
        return"Error: num tiene que ser entero"
            

```
